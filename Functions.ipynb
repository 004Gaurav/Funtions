#1.  What is the difference between a function and a method in Python?

  ~ In Python, both **functions** and **methods** are callable objects, but they have some important distinctions:

1. **Function:**
  - A **function** is a block of reusable code that performs a specific task.
  - It is defined using the `def` keyword and can be called from anywhere in the code.
  - Functions can operate on parameters passed to them but don't need to belong to any particular object or class.
  
  Example:
  ```python
  def add(a, b):
      return a + b
  ```

2. **Method:**
 - A **method** is a function that is associated with an object, typically a class instance (or sometimes a class itself).
 - Methods are defined within a class and are called on instances of that class (or on the class itself in the case of class methods).
 - The first parameter of a method is usually `self`, which refers to the instance of the object.

  Example:
  ```python
  class Calculator:
      def add(self, a, b):
          return a + b
  
  calc = Calculator()
  result = calc.add(3, 4)  # 'add' is a method
  ```

### Key Differences:
- **Functions**: Independent, can be called directly.
- **Methods**: Belong to a class or object and are called on instances or classes.

#4. What is the purpose of the `return` statement in a Python function?

The `return` statement in a Python function serves the following purposes:

1. **Returning a Value**:
   - It sends a result back to the caller of the function. This allows the function to output a value, which can be used or stored elsewhere in the code.
   
   Example:
   ```python
   def add(a, b):
       return a + b

   result = add(2, 3)  # result will be 5
   ```

2. **Exiting the Function Early**:
   - When `return` is executed, it immediately terminates the function, and no further code within the function is executed. This can be useful for stopping a function when a certain condition is met.
   
   Example:
   ```python
   def check_even(number):
       if number % 2 == 0:
           return True
       return False

   print(check_even(4))  # Output: True
   ```

3. **Returning None (default behavior)**:
   - If no value is explicitly returned, the function will return `None` by default.

   Example:
   ```python
   def greet(name):
       print(f"Hello, {name}")

   result = greet("Alice")  # result will be None
   ```

### Key Points:
- The `return` statement allows a function to **output a value**.
- It **stops the function execution** and exits immediately.
- Without a `return` statement, a function implicitly returns `None`.










#5.What are iterators in Python and how do they differ from iterables?

- In Python, **iterables** and **iterators** are related but distinct concepts:

1. **Iterable:**
An **iterable** is any object that can be looped over (e.g., using a `for` loop). It implements the `__iter__()` method, which returns an iterator. Common iterables include lists, strings, and sets.

Example:
```python
my_list = [1, 2, 3]
for item in my_list:  # Iterating through the iterable
    print(item)
```

2. **Iterator:**
An **iterator** is an object that not only provides the next item in a sequence but also keeps track of the current state of iteration. It implements two methods:
 - `__iter__()` (returns the iterator itself)
 - `__next__()` (returns the next item or raises `StopIteration` when exhausted)

Example:
```python
my_list = [1, 2, 3]
iterator = iter(my_list)  # Convert iterable to iterator
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
```

Key Differences:

  - **Iterable**: Can be looped over (e.g., `for item in iterable`), implements `__iter__()`.
  - **Iterator**: Produces items one by one using `__next__()` and maintains iteration state.


#6.  Explain the concept of generators in Python and how they are defined.

In Python, **generators** are a type of iterable that allows you to iterate over a sequence of values without storing the entire sequence in memory. This is especially useful when working with large datasets or infinite sequences, as they generate values on-the-fly as you iterate over them.

### Key Points:
- Generators are defined using functions with the `yield` keyword instead of `return`.
- They generate values one at a time when needed, which means they don’t store the entire sequence in memory.

### How Generators Work:
1. A **generator function** is called like a regular function but it returns a generator object instead of executing the function immediately.
2. Each time the `yield` statement is encountered, the function pauses, returning the yielded value to the caller.
3. The function resumes execution from the point where it paused when the next value is requested.

### Example of a Generator:
```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield the current count
        count += 1

# Using the generator
counter = count_up_to(5)
for num in counter:
    print(num)
```
In this example, the generator `count_up_to(5)` yields numbers from 1 to 5, one by one, as the loop iterates.

### Advantages of Generators:
1. **Memory Efficiency**: Generators only produce values when needed, so they don’t store the entire sequence in memory.
2. **Lazy Evaluation**: Values are computed only when requested, making generators ideal for large or infinite sequences.

### Example of an Infinite Generator:
```python
def infinite_counter():
    count = 0
    while True:
        yield count
        count += 1

# Using the generator
gen = infinite_counter()
for _ in range(5):
    print(next(gen))  # Outputs 0, 1, 2, 3, 4
```
Here, `infinite_counter()` generates an infinite sequence of numbers, producing each number only when requested.

### Summary:
- **Generators** are functions that use `yield` to produce values lazily, one at a time.
- They are more memory-efficient than functions that return large collections because they don’t store all values in memory.
- They are ideal for working with large datasets or infinite sequences due to their lazy evaluation and state retention between iterations.




















#7. What are the advantages of using generators over regular functions?

- Generators offer several advantages over regular functions:

 - **Memory Efficiency**: They generate values one at a time, without storing the entire sequence in memory, making them more memory-efficient for large datasets.
   
 - **Lazy Evaluation**: Values are generated only when needed, improving performance, especially when you don’t need all the data at once.

 - **State Retention**: Generators remember their state between iterations, simplifying iteration logic.

 - **Simpler Code**: Generators often lead to cleaner, more concise code for complex iteration tasks.


#8. What is a lambda function in Python and when is it typically used?

- A **lambda function** in Python is a small, anonymous function defined using the `lambda` keyword. It can have any number of arguments but only one expression, which is evaluated and returned.

### Syntax:
```python
 lambda arguments: expression
```

### Example:
```python
add = lambda x, y: x + y
print(add(2, 3))  # Output: 5
```

### Key Characteristics:

  - **Anonymous**: Lambda functions are often unnamed (not assigned to a variable).
  - **Single Expression**: They can only contain a single expression, which is evaluated and returned automatically.
  - **Used Inline**: Typically used where a small function is needed temporarily.

### When is it used?
1. **Short-term Use**: When a simple function is needed for a short duration, often as an argument to higher-order functions like `map()`, `filter()`, or `sorted()`.
   
   Example with `map()`:
   ```python
   nums = [1, 2, 3, 4]
   squared = map(lambda x: x**2, nums)
   print(list(squared))  # Output: [1, 4, 9, 16]
   ```

2. **Callbacks or Inline Functions**: When a function is needed only for a brief task (e.g., sorting with custom key).
   
   Example with `sorted()`:
   ```python
   pairs = [(1, 'one'), (3, 'three'), (2, 'two')]
   sorted_pairs = sorted(pairs, key=lambda x: x[0])
   print(sorted_pairs)  # Output: [(1, 'one'), (2, 'two'), (3, 'three')]
   ```


#9. Explain the purpose and usage of the `map()` function in Python.

The **`map()`** function in Python is used to apply a specified function to each item in an iterable (like a list, tuple, etc.) and returns an iterator that produces the results.

### Syntax:
```python
map(function, iterable)
```

- **`function`**: The function to apply to each element.
- **`iterable`**: The iterable whose elements will be processed.

### Purpose:
`map()` is useful for transforming each element of an iterable by applying a function to it, allowing for a clean, concise alternative to using loops.

### Examples:

1. **Using a regular function**:
```python
def square(x):
    return x * x

nums = [1, 2, 3, 4]
squared_nums = map(square, nums)
print(list(squared_nums))  # Output: [1, 4, 9, 16]
```

2. **Using a `lambda` function**:
```python
nums = [1, 2, 3, 4]
squared_nums = map(lambda x: x * x, nums)
print(list(squared_nums))  # Output: [1, 4, 9, 16]
```

3. **Mapping with multiple iterables**:
You can pass multiple iterables to `map()`, and the function will process items from them in parallel.
```python
def add(x, y):
    return x + y

 nums1 = [1, 2, 3]
 nums2 = [4, 5, 6]
 sum_nums = map(add, nums1, nums2)
 print(list(sum_nums))  # Output: [5, 7, 9]
```


~Key Points:

- **Returns an iterator**: `map()` returns an iterator, so it’s memory-efficient and needs to be converted to a list or another collection if needed.
- **Functional Programming**: It's a functional programming tool often used to
  apply a function to all elements of a sequence.
- **Parallel processing**: When multiple iterables are passed, `map()` applies the function to the corresponding items in each iterable.




















#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?


* In Python, the **`map()`**, **`reduce()`**, and **`filter()`** functions are all used for functional programming and operate on iterables (e.g., lists, tuples). While they have similar purposes of processing data, they each serve different roles. Here's a breakdown of their differences:

#### 1. **`map()`**:
   - **Purpose**: Applies a function to each item in an iterable and returns a new iterable (map object) with the results.
   - **Output**: An iterator that yields transformed items.
   - **Use Case**: When you need to transform each item in an iterable (e.g., square each number in a list).
   
   - **Syntax**: `map(function, iterable)`

  **Example**:
   ```python
   nums = [1, 2, 3]
   squared = map(lambda x: x**2, nums)
   print(list(squared))  # Output: [1, 4, 9]
   ```

#### 2. **`reduce()`** (from `functools`):
   - **Purpose**: Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value.
   - **Output**: A single result after applying the function cumulatively.
   - **Use Case**: When you need to combine all items in an iterable (e.g., summing numbers, multiplying items).
   
   **Syntax**: `reduce(function, iterable[, initializer])`

   **Example**:
   ```python
   from functools import reduce
   nums = [1, 2, 3, 4]
   result = reduce(lambda x, y: x + y, nums)
   print(result)  # Output: 10 (1 + 2 + 3 + 4)
   ```

---

#### 3. **`filter()`**:
   - **Purpose**: Filters items in an iterable by applying a function that returns `True` or `False`. Only the items where the function returns `True` are included in the result.
   - **Output**: An iterator with only the items that meet the condition (those where the function returns `True`).
   - **Use Case**: When you want to filter out items that don’t meet a certain condition (e.g., finding even numbers in a list).
   
   - **Syntax**: `filter(function, iterable)`

  **Example**:
   ```python
   nums = [1, 2, 3, 4]
   even_nums = filter(lambda x: x % 2 == 0, nums)
   print(list(even_nums))  # Output: [2, 4]
   ```

In summary:
- **`map()`** is for **transforming** data,
- **`reduce()`** is for **combining** data into a single result,
- **`filter()`** is for **selecting** elements based on a condition.

````python
#4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime.")
````

````python
#5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current = 0
        self.next_term = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        else:
            fibonacci_number = self.current
            self.current, self.next_term = self.next_term, self.current + self.next_term
            self.count += 1
            return fibonacci_number

fib = FibonacciIterator(10)
for num in fib:
    print(num)
````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):

    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Removes the newline character at the end of each line
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

file_path = "example.txt"

# Writing a sample file for demonstration
with open(file_path, "w") as f:
    f.write("First line\n")
    f.write("Second line\n")
    f.write("Third line\n")

# Using the generator to read the file
for line in read_file_line_by_line(file_path):
    print(line)
````

````python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 30, 40, 100]

# Using map to convert each Celsius temperature to Fahrenheit
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Output the result
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
````

````python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "This is an example string."

# Using filter to remove vowels
result = ''.join(filter(is_not_vowel, input_string))

# Output the result
print("Original string:", input_string)
print("String without vowels:", result)
````
